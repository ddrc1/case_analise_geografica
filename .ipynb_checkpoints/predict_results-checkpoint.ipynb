## Importação de bibliotecas

```python
import pandas as pd
import numpy as np
import pickle
```

## Carga de dados

```python
df = pd.read_excel('./dados/dados.xlsx')
df
```

### Separação de São Paulo

```python
df_sp = df.query("cidade == 'São Paulo'")
df_sp
```

```python
df_sp.info()
```

## Tratamento de dados

```python
df_sp.query("rendaMedia == '-'")
```

```python
df_sp.drop(df_sp.query("rendaMedia == '-'").index, inplace=True)
```

```python
df_sp
```

## Carga dos modelos

```python
model_reg, model_clf = pickle.load(open("./dados/models.pkl", 'rb'))
```

## Predição

```python
drop_columns = ["faturamento", "potencial", "codigo", "nome", "cidade", "estado"]
faturamento_pred = model_reg.predict(df_sp.drop(columns=drop_columns))
potencial_pred = model_clf.predict(df_sp.drop(columns=drop_columns))
```

```python
df_sp['faturamento'] = faturamento_pred
df_sp['potencial'] = potencial_pred
df_sp
```

```python
df_sp.to_csv('./dados/final.csv', index=False)
```
